```python
#1.1 Ada berapakah total kata unik yang ada pada file tersebut?
isi = []
with open("soal1.txt") as teks:
    for i in teks:
        isi.append(i)

pisah = isi[0].split()

hasil = set(pisah)
len(hasil)
```

```python
#1.2 Kata apakah yang kemunculannya tertinggi ke-6 pada file tersebut?
from collections import Counter

hitung = Counter(pisah)
kata = hitung.most_common(6)
print("Kata Paling Sering: ", kata[-1][0], '\n',"Dengan jumlah kemunculan:",kata[-1][1])
```

```python
import pandas as pd

df = pd.read_csv("soal2_train.csv")

df
```

```python
data = df.values

batas = data.shape[0] * 8 // 10

x = data[:,1]
y = data[:,0].astype(int)

x_train = x[:batas]
x_test = x[batas:]

y_train = y[:batas]
y_test = y[batas:]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

idf = TfidfVectorizer()
x_idf = idf.fit_transform(x_train)
x_test_idf = idf.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(x_idf,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = rfc.predict(x_test_idf)
print(accuracy_score(y_pred,y_test))
```

```python
df2 = pd.read_csv("soal2_test.csv")
data_test = df2.values[:,0]
test_idf = idf.transform(data_test)
```

```python
hasil = rfc.predict(test_idf)

hasil = hasil.tolist()
```

```python
out = pd.DataFrame({"sentimen":hasil})

out
```

```python
out.to_csv("20102153.csv", index = False)
```
