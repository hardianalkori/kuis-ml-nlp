```python
import pandas as pd

# load dataset menggunakan pandas dengan mendefinisikan koma pada nilai yang ada menggunakan tanda koma
df = pd.read_csv("UTS_train.csv", decimal = ",")
df
```

```python
# Merubah nilai pada dataset menjadi float
df = df.astype(float)
```

```python
# melakukan pendefinisian variabel dimana input merupakan x1-x15 dan y nya adalah kolom y
data = df.values

x = data[:,:-1]
y = data[:,-1]
```

```python
# mendesain MLP untuk training
    
from keras.models import Sequential
from keras.layers import Dense, Flatten,Dropout

model = Sequential()

model.add(Dense(100, input_dim = 15, activation = "relu")) #input layer 15 dan 15 neuron
model.add(Dense(200, activation = "relu")) # node/neuron sebanyak 30 dengan fungsi aktivasi relu
model.add(Dense(300, activation = "relu")) # node/neuron sebanyak 45
model.add(Dense(400, activation = "relu")) # node/neuron sebanyak 45
model.add(Flatten()) # melakukan flatting input menjadi 1 dimensi
model.add(Dense(50, activation = "relu"))
model.add(Dense(25, activation = "relu"))
model.add(Dense(3, activation = "relu"))
model.add(Dense(1, activation = "linear")) # menggunakan fungsi aktivasi linear dikarenakan kasusnya adalah regresi
model.compile(loss = "mean_absolute_error",optimizer = "adam", metrics=["mean_absolute_error"]) #mengemas model dengan metrik perhitungan mse dan optimizer menggunakan adam
model.summary() #melihat ringkasan model MLP yang dibuat
```

```python
#melakukan training dengan btach size sebanyak 1 dan epochs sebanyak 100 kali dengan validation 10%
model.fit(x,y, batch_size = 1, epochs = 50, validation_split = 0.1)
```

```python
# load data untuk prediksi
df2 = pd.read_csv("UTS_test_soal.csv", decimal = ',')

df2
```

```python
#merubah tipe data yang ada di dataset menjadi float
df2 = df2.astype(float)
df2
```

```python
#melakukan prediksi
data_pred = df2.values

y_pred = model.predict(data_pred)
```

```python
#hasil prediks
y_pred
```

```python
#menjadikan output prediksi menjadi list untuk membuat dataframe output
konv = y_pred.tolist()
konv = [konv[i][0] for i in range(len(konv))]

konv
```

```python
# menjadikan output menjadi dataframe
hasil = pd.DataFrame({"y":konv})

hasil
```

```python
# export hasil prediksi
hasil.to_csv("20102153.csv", index = False)
```
