```python
#1.1 Ada berapakah total kata unik yang ada pada file tersebut?
isi = []
with open("soal1.txt") as teks:
    for i in teks:
        isi.append(i)

pisah = isi[0].split()

hasil = set(pisah)
len(hasil)
```

```python
#1.2 Kata apakah yang kemunculannya tertinggi ke-6 pada file tersebut?
from collections import Counter

hitung = Counter(pisah)
kata = hitung.most_common(6)
print("Kata Paling Sering: ", kata[-1][0], '\n',"Dengan jumlah kemunculan:",kata[-1][1])
```

```python
#1.3 Carilah jumlah kombinasi huruf yang berakhiran huruf vokal secara regex?
import re

hasil = re.findall(r'\w*[aiueo]\b',isi[0])
hasil
```
