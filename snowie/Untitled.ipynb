```python
import pandas as pd

# load dataset menggunakan pandas dengan mendefinisikan koma pada nilai yang ada menggunakan tanda koma
df = pd.read_csv("UTS_train.csv", decimal = ",")
df
```

```python
df = df.astype(float)
```

```python
data = df.values
batas = data.shape[0] * 8 // 10

x = data[:,:-1]
y = data[:,-1]

x_train = x[:batas]
x_test = x[batas:]

y_train = y[:batas]
y_test = y[batas:]
```

```python
from sklearn.linear_model import Ridge

lr = Ridge()
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error
y_pred = lr.predict(x_test)
print(mean_absolute_error(y_test,y_pred))
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten,Dropout

model = Sequential()

model.add(Dense(15, input_dim = 15, activation = "relu"))
model.add(Dense(30, activation = "relu"))
model.add(Dense(45, activation = "relu"))
model.add(Dense(60, activation = "relu"))
#model.add(Dense(50, activation = "relu"))
model.add(Flatten())
model.add(Dense(1, activation = "linear"))
model.compile(loss = "mean_squared_error",optimizer = "adam", metrics=["mean_squared_error"])
model.summary()
```

```python
model.fit(x,y, batch_size = 1, epochs = 100, validation_split = 0.1)
```
