```python
import pandas as pd

df = pd.read_csv("kalimat.csv")

df
```

```python
x_idf = df["Kalimat"].values
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

idf = TfidfVectorizer()
idf.fit(x_idf)
```

```python
#idf.get_feature_names()
penting = []

def coba(n):
    kalimat = []
    for k,v in enumerate(idf.get_feature_names()):
        c = idf.transform([x_idf[n]])
        bobot = c[0,k]
        kata = v
        if bobot != 0:
            kalimat.append([kata,bobot])
            
    kalimat = sorted(kalimat,key=lambda l:l[1], reverse=True)
    return kalimat[0][0]

for i in range(len(x_idf)):
    penting.append(coba(i))
```

```python
print(penting)
```

```python
df["Rangkuman"] = penting
df
```

```python
df.to_csv("20102153.csv",index = False)
```
