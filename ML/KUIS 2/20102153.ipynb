```python
import cv2
from keras.models import Sequential
from keras.layers import Dense
```

```python
import glob

x = []
y = []

for nama_file in glob.glob('0/*.jpg'):
  gambar = cv2.imread(nama_file)
  gambar = cv2.resize(gambar,(64,64))/255
  x.append(gambar)
  y.append(0)

for nama_file in glob.glob('1/*.jpg'):
  gambar = cv2.imread(nama_file)
  gambar = cv2.resize(gambar,(64,64))/255
  x.append(gambar)
  y.append(1)

for nama_file in glob.glob('2/*.jpg'):
  gambar = cv2.imread(nama_file)
  gambar = cv2.resize(gambar,(64,64))/255
  x.append(gambar)
  y.append(2)
```

```python
import random
import numpy as np

z =list(zip(x,y))
random.shuffle(z)
x,y = zip(*z)
x = np.array(x)
y = np.array(y)
```

```python
x.shape
```

```python
from keras.layers import Flatten,Conv2D, MaxPooling2D,Dense,Dropout
model = Sequential()
model.add(Conv2D(8,(3,3),input_shape=(64,64,3),activation = 'relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(16,(3,3),activation = 'relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(32,(3,3),activation = 'relu'))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
model.add(Dense(30,activation = 'relu'))
model.add(Dense(15,activation = 'relu'))
model.add(Dense(3,activation = 'softmax'))
model.compile(optimizer = 'adam',loss='sparse_categorical_crossentropy',metrics=["accuracy"])
model.summary()



```

```python
model.fit(x,y, batch_size = 64,epochs = 50,validation_split = 0.2)
```

```python
model.save_weights("hasil5.h5")
```
