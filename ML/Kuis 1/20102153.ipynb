```python
import pandas as pd

df = pd.read_csv("train.csv")

df.head()
```

```python
test = pd.read_csv("test_soal.csv")
test.head()
```

```python
print(df.shape)
print(test.shape)
```

```python
df.info()
```

```python
train = df.values
batas = train.shape[0] * 8 // 10

x = train[:,:-1]
y = train[:,-1]

x_train = x[:batas]
x_test = x[batas:]
y_train = y[:batas]
y_test = y[batas:]

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter = 30000, solver = 'lbfgs')
lr.fit(x_train,y_train)
lr.score(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = lr.predict(x_test)
print(accuracy_score(y_pred,y_test))
```

```python
x_test = test.values
hasil_pred = lr.predict(x_test)
```

```python
hasil = pd.DataFrame({"status": hasil_pred.ravel()}).astype(int)

hasil.to_csv("2010215.csv", index = False)
```

```python
from sklearn.model_selection import cross_val_score

crosval = cross_val_score(LogisticRegression(), x_train,y_train, cv=10)

print(crosval.ravel().mean())
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train,y_train)
rf.score(x_train,y_train)
```

```python


crosval = cross_val_score(RandomForestClassifier(), x_train,y_train, cv=10)

print(crosval.ravel().mean())
```
